# bertrand2025/figure1
Plot figure 1 from Bertrand et al. 2025, "Increasing wintertime cloud opacity increases surface longwave radiation at a long-term  Arctic observatory".

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import xarray as xr
import statsmodels.api as sm
```

# 1.1 load data

```python
declim = lambda ds: ds.groupby('time.month') - ds.groupby('time.month').mean()
ds_submos = xr.open_dataset('monthly_mean_metcldrad_features.nc')
arm_t2m = xr.open_dataset('ARM_NSA_C1_surface_air_temperature.nc')
gml_t2m = xr.open_dataset('NOAA_GML_BRW_ABO_surface_air_temperature.nc')
ds_hists = xr.open_dataset('surface_longwave_flux_histograms.nc')
#absolute temperatures
t2m_mons = { }

#add in GML and ARM met data
t2m_mons['NOAA BRW GML'] = gml_t2m
t2m_mons['ARM NSA C1'] = arm_t2m

#monthly temperature anomalies (remove seasonal cycle)
t2m_mon_anoms = { }
for label,t2m_mon in t2m_mons.items():
    t2m_mon_anom = t2m_mon.groupby('time.month') - t2m_mon.groupby('time.month').mean()
    t2m_mon_anoms[label] = t2m_mon_anom
```

## 1.2 plot

```python
#compute all 4 regressions, saving results
def do_regression(predictand,predictor,data_submos):
    '''Ordinary least-squares regression'''
    # do DJFM-only OLS regression
    ds = data_submos
    Y, X = xr.align(ds[predictand].dropna('time'),ds[predictor].dropna('time'))
    Y, X = Y.data, X.data
    X = sm.add_constant(X)
    olsr_djfm_res = sm.OLS(Y,X, missing='drop').fit()
    return olsr_djfm_res
```

```python
def get_tmask(bounds,dax_std):
    '''get temporal mask indicating months with a temperature anomaly in a range'''
    thresh_inner, thresh_outer = bounds
    thresh = 0.5*(thresh_inner+thresh_outer)
    #get subset mask
    if thresh < 0: #left tail for neg
        tmask = (thresh_outer<=dax_std) & (dax_std<=thresh_inner)
    elif thresh > 0: #right tail for pos
        tmask = (thresh_outer>=dax_std) & (dax_std>=thresh_inner)
    elif thresh == 0: #all points for zero 
        tmask = xr.ones_like(dax_std,dtype=bool)
    return tmask

def decade_to_datetime(decade,t0):
    dt = pd.to_timedelta(decade*(1e9*60*60*24*365.25*10),unit='ns') #decades to nanoseconds
    return dt+pd.to_datetime(t0)

def my_precision(x, n):
    '''30897226'''
    return '{:.{}f}'.format(x, n)
```

```python
decade = ds_submos['decade']
keep_months = [12,1,2,3]
sub_mos = lambda da: da.sel(time=da['time.month'].isin(keep_months))
```

```python
%%time
fig, axes = plt.subplots(2,2,figsize=(10,5),sharey=False,dpi=300,height_ratios=[2,1])

pretty_names = {'lwu':'Longwave up','lwd':'Longwave down','lwn':'Net longwave',
                'Temp_Air':'Surface air temperature anomaly'}
#tol vibrant
colors = {'lwu': '#0077BB', 'lwd': '#EE7733', 'lwn':'#009988'}

### REGRESSIONS - TOP ROW ###
for predictor,ax in zip(['decade','Temp_Air'],axes[0,:]):
    for predictand in ['lwu','lwd','lwn']:
        #scatterplot
        yi,xi = xr.align(ds_submos[predictand].dropna('time'),ds_submos[predictor].dropna('time'))
        xp = decade_to_datetime(xi,decade.time.data[0]) if predictor == 'decade' else xi
        ax.scatter(xp,yi,marker='.',color=colors[predictand],alpha=0.4)
        #trendline
        res = do_regression(predictand,predictor,ds_submos)
        xx = np.linspace(xi.min(),xi.max())
        ycap = xx*res.params[1]+res.params[0]
        print(f'{predictor} vs. {predictand}: {res.params[1]:.2f}')
        #trendline 95% confidence band
        cov = res.cov_params()
        var_ycap = cov[0,0] + (xx**2)*cov[1,1] - 2*xx*cov[0,1]
        err_ycap = 2*np.sqrt(var_ycap)
        #plot trendline
        xp = decade_to_datetime(xx,decade.time.data[0]) if predictor == 'decade' else xx
        slope_units=f"{yi.attrs['units']}/{xi.attrs['units']}"
        min_slope, max_slope = res.conf_int(0.05)[1]
        min_slope = my_precision(min_slope,2-int(np.log10(np.abs(min_slope))))
        max_slope = my_precision(max_slope,2-int(np.log10(np.abs(max_slope))))
        ax.plot(xp,ycap,color=colors[predictand],label='{:}: {:}-{:} {:}'.format(pretty_names[predictand],
                                                                                 min_slope,max_slope,
                                                                                 slope_units))
        #plot confidence band
        ax.fill_between(xp, ycap-err_ycap, ycap+err_ycap, color=colors[predictand], alpha=0.2)
        if predictor != 'decade': ax.set_xlabel(f'{pretty_names[predictor]} [{xi.attrs["units"]}]')
        #if predictand == 'lwn': break
    legend = ax.legend(loc='lower center',bbox_to_anchor=[0.5,-0.5])
    texts  = legend.get_texts()
    #[text.set_weight('heavy') for text in texts[:3]]
    ax.set_ylabel('Surface longwave flux anomaly [W/m$^2$]')
#move temperature trendline to come first so the legends align
ax = axes[0][0]
handles, labels = ax.get_legend_handles_labels()
order = [0,1,2] #[3,0,1,2]
legend = ax.legend([handles[idx] for idx in order],[labels[idx] for idx in order],
                   loc='lower center',bbox_to_anchor=[0.5,-0.5])

### HISTOGRAMS - BOTTOM ROW ###
#### DATA PREP ####
ds  = ds_hists
# flux and datastream, site for temp
day = ds.hist.sel(datastream='nsaqcrad1longC1.c2c1', term='LWnet')
dax = t2m_mon_anoms['ARM NSA C1']['Temp_Air']
#select months of interest
day = day.sel(time=day['time.month'].isin(keep_months))
dax = dax.sel(time=dax['time.month'].isin(keep_months))
#get bin bounds
bin_min = day.bin_min
bin_max = day.bin_max
bins = [*list(bin_min),bin_max[-1]]
bin_mids = 0.5*(bin_min+bin_max)
#drop missing values
mask = ~np.isnan(dax) & ~np.isnan(day).all(dim='bin') & (day.sum(dim='bin')>0)
dax,day = dax.where(mask,drop=True), day.where(mask,drop=True)
#standardize temperature
dax_std = (dax-dax.mean())/dax.std()

#criteria for temporal and temperature subsets
win =       40   #number of DJFM months to include
cold_min = -0.5    #standardized anomaly bounds
cold_max = -5 
warm_min =  0.5
warm_max =  5
#insert of +4K sim result
ds_p4k  = xr.open_dataset('14-4-nsa_rrtm_minutely_v2_plus4k_surf_to2015.nc')
ds_no4k = xr.open_dataset('14-4-nsa_rrtm_minutely_v2_vanilla_surf_all.nc')
ds_p4k,ds_no4k = xr.align(ds_p4k,ds_no4k)
counts_p4k, _  = np.histogram(-ds_p4k.net_flux,bins=bins)
counts_no4k, _ = np.histogram(-ds_no4k.net_flux,bins=bins)
counts_diff4k = counts_p4k/counts_p4k.sum()-counts_no4k/counts_no4k.sum()
counts_diff4k = xr.DataArray(data=counts_diff4k,dims='bin',coords={'bin':bin_mids})
#methods to return the proper subsets
slicers = {'decade':  {'initial': lambda da: da.isel(time=slice(0,win)),
                       #'all':     lambda da: da,
                       'warming alone': lambda da: da.isel(time=slice(0,win))+(counts_diff4k*da.isel(time=slice(0,win)).sum('bin')),
                       'final':   lambda da: da.isel(time=slice(-win,None))},
           'Temp_Air':{
                       'cold':    lambda da: da.sel(time=get_tmask([cold_min,cold_max],dax_std)),
                       'all':     lambda da: da,
                       'warm':    lambda da: da.sel(time=get_tmask([warm_min,warm_max],dax_std))}
          }

#plot properties
colors = {'initial':'tab:blue','final':'tab:red','warming alone':'palevioletred',
          'cold':'tab:blue','warm':'tab:red','all':'tab:grey'}
linestyles = {'initial':'-','final':'-','warming alone':':',
          'cold':'-','warm':'-','all':'-'}
xlabels = {'LWnet':'Surface net longwave [W/m$^2$]','LWdn':'Downwelling longwave [W/m$^2$]',
           'LWup':'Upwelling longwave [W/m$^2$]'}
for predictor,ax in zip(['decade','Temp_Air'],axes[1,:]):
    for name,slicer in slicers[predictor].items():
        print(predictor)
        #subset
        day_sub  = slicer(day)
        icounts  = day_sub.sum(dim='time').astype(float)

        #get plot label
        if predictor == 'decade':
            label = f"{day_sub['time.year'].min().data}-{day_sub['time.year'].max().data}"# (N={day_sub.sizes['time']})"
            if name == 'warming alone':
                label = 'Warming and water vapor alone (+4K)'
        elif predictor == 'Temp_Air':
            label = name
            if name == 'cold':
                label = f"$T^\prime\leq{cold_min:.1f}\sigma$"
            elif name == 'warm':
                label = f"$T^\prime\geq{warm_min:.1f}\sigma$"
            #label = label + f" (N={day_sub.sizes['time']})"

        extra_kws = {'lw':2,'dashes':(3, 1),'alpha':0.9} if name=='warming alone' else {'lw':1.5}
        #plot histogram
        ax.plot(bin_mids,icounts/icounts.sum(),zorder=-100,
                label=label,
                linestyle=linestyles[name],
                color=colors[name],**extra_kws)
    
    Ti = list(slicers[predictor].values())[0](dax).mean().data
    Tf = list(slicers[predictor].values())[-1](dax).mean().data
    Fi = list(slicers[predictor].values())[0](day).mean().data
    Ff = list(slicers[predictor].values())[-1](day).mean().data
    ax.text(0.7,0.85,f"$\Delta T=${Tf-Ti:.2f} K ",transform=ax.transAxes)
    #ax.text(0.7,0.8,f"$\Delta F_N=${Ff-Fi:.2f} K ",transform=ax.transAxes)
    ax.set_xlabel(xlabels[day.term.item()])
    ax.set_ylim(ymin=0)
    ax.legend(bbox_to_anchor=(0.5,-0.35),loc='upper center')
    ax.set_ylabel('Frequency')
    if day.term == 'LWnet':
        ax.set_xlim([-100,50])
    else:
        ax.set_xlim([100,350])

#add labels
letter = iter('abcdefghijklmnopqrstuvwxyz')
for ax in axes.flatten():
    ax.text(0.0,1.05,next(letter),weight='bold',size='large',transform=ax.transAxes)

fig.subplots_adjust(top = 1.1, bottom=0.01, hspace=0.8, wspace=0.3)
#dLWNdT = ((-ds_p4k.net_flux).mean()-(-ds_no4k.net_flux).mean())/4 #W/m2/K
#print(f'feedback implied by +4K const-RH minutely flux calculations: {dLWNdT.data:.2f} W/m$^2$/K')
#fig.savefig('67-2-fig01_bertrand2025_R2R_tweaks.pdf',bbox_inches='tight')
```
